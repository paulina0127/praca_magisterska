```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
```

### Porównanie wyników modeli bazowych i zoptymalizowanych

#### Regresja logistyczna

```python
# Metryki
metrics = ["Dokładność", "Precyzja", "Czułość", "F1-score"]
x = np.arange(len(metrics))

# Wyniki modeli
base_min = [0.66, 0.45, 0.25, 0.35]
base_max = [0.84, 0.93, 0.67, 0.70]
base_avg = [0.76, 0.68, 0.46, 0.54]

opt_min = [0.57, 0.35, 0.54, 0.49]
opt_max = [0.82, 0.69, 0.90, 0.74]
opt_avg = [0.69, 0.52, 0.73, 0.60]

width = 0.35

fig, ax = plt.subplots(figsize=(10, 5))

# Bazowy
ax.bar(
    x - width / 2, base_min, width, color="#1f78b4", label="Min (model bazowy)"
)
ax.bar(
    x - width / 2,
    [maxv - minv for minv, maxv in zip(base_min, base_max)],
    width,
    bottom=base_min,
    color="#87ceeb",
    label="Max (model bazowy)",
)
for i in range(len(x)):
    ax.hlines(
        base_avg[i],
        x[i] - width / 2 - 0.1,
        x[i] - width / 2 + 0.1,
        colors="black",
        linewidth=2, 
    )

# Optymalny
ax.bar(
    x + width / 2, opt_min, width, color="#33a02c", label="Min (model zoptymalizowany)"
)
ax.bar(
    x + width / 2,
    [maxv - minv for minv, maxv in zip(opt_min, opt_max)],
    width,
    bottom=opt_min,
    color="#a6d854",
    label="Max (model zoptymalizowany)",
)
for i in range(len(x)):
    ax.hlines(
        opt_avg[i],
        x[i] + width / 2 - 0.1,
        x[i] + width / 2 + 0.1,
        colors="black",
        linewidth=2,
    )

legend_elements = [
    Patch(facecolor='#1f78b4', label='Min (model bazowy)'),
    Patch(facecolor='#87ceeb', label='Max (model bazowy)'),

    Patch(facecolor='#33a02c', label='Min (model zoptymalizowany)'),
    Patch(facecolor='#a6d854', label='Max (model zoptymalizowany)'),

    Line2D([0], [0], color='black', marker='_', markersize=15, label='Średnia wartość', linewidth=0, markeredgewidth=2)
]

ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.set_ylabel("Wartość")
ax.legend(loc="upper left", bbox_to_anchor=(1.0, 1.0), handles=legend_elements)
ax.set_ylim(0, 1.05)
ax.set_yticks(np.arange(0, 1.1, 0.1))
ax.grid(axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```

### Porównanie skuteczności modeli bazowych

```python
data = {
    "Model": [
        "Regresja logistyczna",
        "Regresja logistyczna",
        "Regresja logistyczna",
        "Regresja logistyczna",
        "Las losowy",
        "Las losowy",
        "Las losowy",
        "Las losowy",
        "K-nearest neighbors",
        "K-nearest neighbors",
        "K-nearest neighbors",
        "K-nearest neighbors",
    ],
    "Metryka": ["Dokładność", "Precyzja", "Czułość", "F1-score"] * 3,
    "Wartość": [
        # Regresja logistyczna
        0.76,
        0.68,
        0.46,
        0.54,
        # Las losowy
        0.76,
        0.63,
        0.57,
        0.60,
        # KNN
        0.73,
        0.59,
        0.49,
        0.53,
    ],
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 5))

palette = [
    "#ff9896",
    "#17becf",
    "#FFDE21",
]

ax = sns.barplot(x="Metryka", y="Wartość", hue="Model", data=df, palette=palette)

for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", label_type="edge", padding=3)

plt.ylim(0, 1)
plt.legend()
plt.tight_layout()
plt.show()
```
